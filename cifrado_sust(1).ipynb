```python
#FUNCIONES
#primeramente sin funciones quitaba acentos y caracteres especiales, pero demoraba demasiado, y me di cuenta que
#hacia el anterior proceso para TODO el texto demasiado grande. así que mejor creé funciones para primero que
#nada eliminar las palabras repetidas y con base a las no repetidas trabajar ya que se reducen demasiado
#así le tenemos que quitar acentos y caracteres a 20,000 palabras y no a mucho más de 100,000. 
#y así se reduce drasticamente el tiempo de compilacion
def eliminarRepetidas(dat):
    palabras=""                         #creamos variable para almacenar las palabras
    datos2=dat.split()                  #separa
    for i in datos2:                    #una vez limpio "datos2" de caracteres especiales, creamos en el diccionario
      if not (i in palabras):           #if para que no se repitan las palabras almacenadas en el diccionario
          palabras=palabras+" "+i       #agregamos un espacio para que no quede todo encimado y podamos crear el dicc 
    return palabras                     #separando por espacios, mas adelante

def sinAcentos(dat):                    #funcion que elimine acentos
    dat = dat.replace("á","a")
    dat = dat.replace("é","e")
    dat = dat.replace("í","i")
    dat = dat.replace("ó","o")
    dat = dat.replace("ú","u")
    dat = dat.replace('"',"")
    return dat

def sinCaracEsp(dat):                   #funcion elimina caracteres especiales
    for c in dat:                       #for para REMPLAZAR caracteres especiales que no necesitamos para el diccionario
      if c in "…—––!/\()0123456789,•.&-:;?!*‘▶´'´’‘“”«»¶,=%+©−'\_`>#,\7\01":  
        dat = dat.replace(c,"")         #remplazar por ""
      else:                             
        dat=dat+c                       #obtener todo aquello que no tenga caracteres especiales, si no estariamos perdiendo palabras
    return dat

#ABRIR ARCHIVO
arch=open("FrancesTexto.txt", "r")
datos=arch.read()                 #lee
datos=datos.lower()               #minusculas

datos2=eliminarRepetidas(datos)         #creamos datos2
#print("sin palabras repetidas",datos2)
datos2=sinCaracEsp(datos2)
#print("sin caracteres especiales es",datos2)
datos2=sinAcentos(datos2)
arch.close()                      #cierra

datos2=datos2.split()              #separa para crear el diccionario
dicc=[]                           #creamos variables para el diccionario
for i in datos2:                  #una vez limpio "datos2" de caracteres especiales, creamos en el diccionario
  if not (i in dicc):             #if para que no se repitan las palabras cada que se recorra datos2 **?
      dicc.append(i)              #agregamos
dicc.sort(key=len)                       #toma una lista y la ordena con la llave "len" de palabras de menor a mayor caracter

print("El diccionario es:",dicc)
print("número de palabras en el diccionario es:",len(dicc))          #obtenemos el numero de palabras

arch3 = open('words.txt', 'w')
for e in dicc:
  arch3.write(e+" ")
arch3.close()

#print("texto es",datos2)       ['il', 'n’y', 'a', 'qu’un', 'problème', 'philosophique'...
```
